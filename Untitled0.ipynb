```python
import numpy as np

# Arrays (same as before)
years_arr = np.array([1965, 1966, 1967, 1968, 1969])
murder_arr = np.array([836, 882, 996, 1185, 1324])
pop_arr = np.array([18073000, 18258000, 18336000, 18113000, 18321000])

# Murder rates
murder_rates = (murder_arr / pop_arr) * 100000
print("Murder rates:", murder_rates)

# Find index of highest rate
max_index = np.argmax(murder_rates)

# Get year and value
year_highest = years_arr[max_index]
highest_rate = murder_rates[max_index]

print("Year with highest murder rate:", year_highest)
print("Highest murder rate:", round(highest_rate, 2))
```

```python
# Step 1: Import libraries
import pandas as pd
from google.colab import files

# Step 2: Upload the CSV file from your computer
print("Please select the 'crime.csv' file to upload:")
uploaded = files.upload()  # This opens a file picker

# Step 3: Load the dataset into a DataFrame
# The uploaded file's name is the key in the 'uploaded' dictionary
filename = list(uploaded.keys())[0]
crime_df = pd.read_csv(filename)

# Step 4: Basic inspection
print("\nFirst 5 rows:")
print(crime_df.head())

print("\nDataset info:")
crime_df.info()

print("\nSummary statistics:")
print(crime_df.describe())
```

```python
# Show the last 10 rows
print("Last 10 rows of the dataset:")
print(crime_df.tail(10))

# Check for missing values in each column
missing_values = crime_df.isnull().sum()
print("\nMissing values per column:")
print(missing_values)

# Identify the column(s) with the most missing data
max_missing = missing_values.max()
columns_most_missing = missing_values[missing_values == max_missing].index.tolist()
print(f"\nColumn(s) with the most missing data ({max_missing} missing values): {columns_most_missing}")
```

```python
recent_crimes = crime_df[crime_df['Year'] > 2000]
print("Recent crimes shape:", recent_crimes.shape)

# Select only Murder and Robbery columns for years with high assault (> 60000)
high_assault = crime_df.loc[crime_df['Assault'] > 60000, ['Year', 'Murder', 'Robbery']]
print("High assault years (Murder, Robbery):\n", high_assault)
```

```python
# Filter rows where CarTheft > 200,000 and select specific columns
high_cartheft = crime_df.loc[crime_df['CarTheft'] > 200000, ['Year', 'Population', 'CarTheft']]

print("Years with CarTheft > 200,000:\n", high_cartheft)
```

```python
crime_df['decade'] = (crime_df['Year'] // 10) * 10

# Group by decade and compute mean for numeric columns
decade_stats = crime_df.groupby('decade')[['Murder', 'Rape', 'Robbery', 'CarTheft']].mean()
print("Average crimes by decade:")
print(decade_stats)
```

```python
# Create 'decade' column (if not already created)
crime_df['decade'] = (crime_df['Year'] // 10) * 10

# Group by decade and compute median for numeric columns including Burglary
decade_median = crime_df.groupby('decade')[['Murder', 'Rape', 'Robbery', 'Burglary', 'CarTheft']].median()
print("Median crimes by decade:\n", decade_median)

# Find the decade with the highest median Burglary
max_burglary_index = decade_median['Burglary'].idxmax()
max_burglary_value = decade_median['Burglary'].max()
print(f"\nDecade with highest median Burglary: {max_burglary_index}s ({max_burglary_value} cases)")
```

```python
# Reload data (if needed)
df = pd.read_csv('crime.csv')

# Check nulls
print("Null values per column:\n", df.isnull().sum())

# If any nulls exist, drop rows with any null (or fill with 0)
df_clean = df.dropna()
print("Shape after dropping nulls:", df_clean.shape)

# Sort by Robbery descending
sorted_df = df_clean.sort_values('Robbery', ascending=False)
print("Top 5 years by Robbery:\n", sorted_df[['Year', 'Robbery']].head())
```

```python
# Sort by Murder in ascending order
sorted_murder = df_clean.sort_values('Murder', ascending=True)

# Display first 5 rows
print("Top 5 years with lowest Murder counts:\n", sorted_murder[['Year', 'Murder']].head())

# Reset index
sorted_murder = sorted_murder.reset_index(drop=True)
print("\nDataFrame after resetting index:")
print(sorted_murder[['Year', 'Murder']].head())
```
